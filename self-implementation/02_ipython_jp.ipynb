```python
# duck typing or catching exceptions
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:
        return False
```

```python
isiterable('a string')
```

## Datetime

```python
from datetime import datetime, date, time

dt = datetime(2011, 10, 30, 20, 30, 21)

dt.month, dt.minute
```

```python
dt.date(), dt.time()
```

```python
dt.strftime("%Y-%m-%d %H:%M")
```

```python
datetime.strptime('20091031', "%Y%m%d")
datetime(2009, 10, 21, 0, 0)
```

```python
# Datetime objects are immutable
# replace method returns new datetime object
dt_hour = dt.replace(minute= 0, second= 0)
dt_hour
```

```python
# timedelta object
# Used for amount of time
dt2 = datetime(2011, 11, 15, 22, 30)
delta = dt2 - dt
delta
```

```python
dt + delta
```
